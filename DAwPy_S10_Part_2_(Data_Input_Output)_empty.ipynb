# SQL Connections


```python
pip install sqlalchemy
```

```python
from sqlalchemy import create_engine # py ile sql arasindaki bag create_engine
```

```python
temp_db = create_engine("sqlite:///:memory:") # ram de gecici olarak irtibat saglayacak db yapar
```

```python
temp_db.table_names() # yeni olusturudugumz icin ici bos
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
```

```python
tables
```

```python
tables[6] # yukarda bir suru tablo cikti biz bir tanesini cektik
```

```python
tables[6].to_sql(name= "populations",con = temp_db)
```

```python
temp_db.table_names()
```

```python
pd.read_sql(sql= "populations", con = temp_db)
```

```python
pd.read_sql_query(sql="SELECT Country,Population FROM populations", con = temp_db)
```
