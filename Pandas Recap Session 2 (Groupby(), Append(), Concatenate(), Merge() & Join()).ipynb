The join() is a Pandas library function used to join or concatenate different DataFrames. For example, the join() function joins columns with other DataFrame either on an index or a key column.

Joining is a convenient method for **combining the columns of two potentially differently-indexed** DataFrames into a single result DataFrame.

Among the three DataFrame operations, join allows **the lowest level of control.**

It will combine all the columns from the two tables, with the common columns renamed with the defined **``lsuffix``** and **``rsuffix``**.

The way that rows from the two tables are combined is defined by **``how``**.

The last technique you'll learn is **``merge()``**. You can use merge() **any time** you want to do database-like join operations. It's the **most flexible** of the three operations you've learnt so far [SOURCE](https://realpython.com/pandas-merge-join-and-concat/).

- **What data do you want to join?** â€“ This will be determined by your ``"right"`` and ``"left"`` datasets.<br>
- **How do you want to join them?** â€“ You have different ways to join two datasets â€“ ``"left"``, ``"right"``, ``"full"``, ``"inner"``. It's up to you depending on your final output requirements.
- **What do you want to join them with?** â€“ You also need to tell Pandas how your two datasets are related. What is the ``common column between the two``?

For this part, we have two dataframes â€“ ``product`` and ``customer``. The product dataframe contains product details like Product_ID, Product_name, Category, Price, and Seller_City. The customer dataframe contains details like id, name, age, Product_ID, Purchased_Product, and City.

**Our task is to use our joining skills and generate meaningful information from the data.** 

```python
product = pd.DataFrame({
                    'Product_ID': [101,102,103,104,105,106,107],
                    'Product_name': ['Watch','Bag','Shoes','Smartphone','Books','Oil','Laptop'],
                    'Category': ['Fashion','Fashion','Fashion','Electronics','Study','Grocery','Electronics'],
                    'Price': [299.0,1350.50,2999.0,14999.0,145.0,110.0,79999.0],
                    'Seller_City': ['Delhi','Mumbai','Chennai','Kolkata','Delhi','Chennai','Bengalore']
})
```

```python
customer = pd.DataFrame({
                    'customer_id': [1,2,3,4,5,6,7,8,9],
                    'name': ['Olivia','Aditya','Cory','Isabell','Dominic','Tyler','Samuel','Daniel','Jeremy'],
                    'age': [20,25,15,10,30,65,35,18,23],
                    'Product_ID': [101,0,106,0,103,104,0,0,107],
                    'Purchased_Product': ['Watch','NA','Oil','NA','Shoes','Smartphone','NA','NA','Laptop'],
                    'City': ['Mumbai','Delhi','Bangalore','Chennai','Chennai','Delhi','Kolkata','Delhi','Mumbai']
})
```

```python
product
```

```python
customer
```

**``NOTE:``** By default, the ``merge()`` function performs an inner join. It takes both the dataframes as arguments and the name of the column on which the join has to be performed.

**``TASK :`` Let's assume that we want to know about all the products sold online and who purchased them. Here applying for inner join, we should focus on the ``'Product_ID'`` column where is common at the ``product and customer dataframes``.**

```python
pd.merge(product, customer, on='Product_ID')
```

However, sometimes or in real-world cases, the column names are different in two dataframes. Then, we have to explicitly mention both column names.

In such a scenerio ``'left_on'`` and ``'right_on'`` are two arguments through which we can achieve this. ``left_on`` is the name of the key in the left dataframe and ``right_on`` in the right dataframe:

```python
pd.merge(product, customer, left_on='Product_name', right_on='Purchased_Product')
```

**``TASK:`` Our supervisor now wants more details about the products sold. S/he wants to know about all the products sold by the seller to the same city, in which the seller and customer both belong to the same city.**

In such a scenerio, we again have to perform an inner join on both ``Product_ID`` and ``Seller_City`` of product and ``Product_ID`` and ``City`` columns of the customer dataframe.

```python
pd.merge(product, customer, how='inner', left_on=['Product_ID', 'Seller_City'], right_on=['Product_ID', 'City'])
```

**``TASK:`` We are requested to find all the products that are not sold and all the customers who didnâ€™t purchase anything from us.**

In such a scenerio, we can use Full Join, also known as Full Outer Join, which returns all those records where can be either in the left or right dataframe. On the other hand, when rows in both the dataframes do not match, the resulting dataframe will have ``NaN`` for every column of the dataframe that lacks a matching row.

```python
pd.merge(product, customer, on='Product_ID', how='outer')
```

Notice that all the non-matching rows of both the dataframes have ``NaN`` values for the columns of each dataframe. In this case you can be stucked.<br>
Thanks to Pandas which provides us a fantastic solution: the ``indicator`` argument. Assigning it as True, we will have a new column of ``_merge`` in the outcome and you will be familiar with which row belongs to which dataframe:

```python
products_unsold = pd.merge(product, customer, on='Product_ID', how='outer', indicator=True)
products_unsold
```

```python
products_unsold[~(products_unsold["Product_ID"].isna()) & (products_unsold["Purchased_Product"].isna())]
```

```python
pd.merge(product, customer, on='Product_ID', how='right')
```

**``TASK:`` Let us assume that our supervisor wants information about only those customers who bought something from us.** 

In such a scenerio, we can use Left Join, also known as Left Outer Join, which returns a dataframe containing all the rows of the left dataframe.

In ``Left Join()``, all the non-matching rows of the left dataframe contain ``NaN`` for the columns in the right dataframe. It is simply an inner join plus all the non-matching rows of the left dataframe filled with NaN for columns of the right dataframe.

With respect to our task, all the unsold products will contain NaN for the columns belonging to the customer dataframe.

```python
pd.merge(product, customer, on='Product_ID', how='left')
```

**``TASK:`` Let's assume that we are requested to create a table of information about the products customers bought.**

In such a scenerio, we can use Right Join, also known as Right Outer Join, which returns a dataframe containing all the rows of the right dataframe and only those of the left dataframe that are common in both.

```python
pd.merge(product, customer, on='Product_ID', how='right')
```

## To Sum Up (Difference(s) between merge() and concat() in pandas)

**At a high level:**

- .concat() simply stacks multiple DataFrame together either vertically, or stitches horizontally after aligning on index
- .merge() first aligns two DataFrame' selected common column(s) or index, and then pick up the remaining columns from the aligned rows of each DataFrame.

**More specifically, .concat():**

- Is a top-level pandas function
- Combines two or more pandas DataFrame vertically or horizontally
- Aligns only on the index when combining horizontally
- Errors when any of the DataFrame contains a duplicate index.
- Defaults to outer join with the option for inner join

**And .merge():**

- Exists both as a top-level pandas function and a DataFrame method (as of pandas 1.0)
- Combines exactly two DataFrame horizontally
- Aligns the calling DataFrame's column(s) or index with the other DataFrame's column(s) or index
- Handles duplicate values on the joining columns or index by performing a ``cartesian product`` [What is Cartesian Product in Pandas merge](https://stackoverflow.com/questions/13269890/cartesian-product-in-pandas)
- Defaults to inner join with options for left, outer, and right

Note that when performing pd.merge(left, right), if left has two rows containing the same values from the joining columns or - index, each row will combine with right's corresponding row(s) resulting in a cartesian product. On the other hand, if .concat() is used to combine columns, we need to make sure no duplicated index exists in either DataFrame.

**Practically speaking:**

- Consider .concat() first when combining homogeneous DataFrame, while consider .merge() first when combining complementary DataFrame.
- If need to merge vertically, go with .concat(). If need to merge horizontally via columns, go with .merge(), which by default merge on the columns in common.

[SOURCE](https://stackoverflow.com/questions/38256104/differences-between-merge-and-concat-in-pandas)

```python
# Cartesion Example

df1 = pd.DataFrame({'key':[1, 1], 'col1':[1, 2]})
df2 = pd.DataFrame({'key':[1, 1, 1], 'col3':[5, 6, 7]})

pd.merge(df1, df2, on='key')[['col1', 'col3']]
```

## Pandas Merge Tip

When you merge two DataFrames, there are often columns you don't want to merge in either dataset. For example, saying that you have two DataFrames with 100 columns distinct columns each, but you only care about 3 columns from each one.

Instead of joining two entire DataFrames together, youâ€™ll only join a subset of columns together. See the code below.

**This join brings together the entire DataFame**

- ``df.merge(df2)``

**This join only brings together a subset of columns**

"col1" is the key in this case

- ``df[["col1", "col2"]].merge(df2[["col1", "col3"]], on="col1")``
