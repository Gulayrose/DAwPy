q: Find the Total Number of Each Product Sold

```python
##bu sorunun cevabi aslinda yukarda bulundu simdi baska yontemler ile bulacagiz sonuclari
```

```python
df.groupby("Product")['Quantity Ordered'].transform(lambda x:np.sum(x))
```

```python
df.groupby("Product")['Quantity Ordered'].transform('sum')
```

```python
df.groupby("Product")['Quantity Ordered'].transform(lambda x:x.sum())
```

```python
##Gives Nan values
##df["Total Revenue of Each Product"]= df.groupby("Product")['Quantity Ordered'].apply(lambda x:np.sum(x))
##burda groupby yapilacaksa transform kullanilir. eger bir seye assayn edeceksek apply da nan veriyo
df["Total Revenue of Each Product"]= df.groupby("Product")['Quantity Ordered'].transform(lambda x:np.sum(x))
df
```

q: Find the Total Revenue of Each Product Sold

```python
df['Price Each']*df["Total Revenue of Each Product"] #bu ikisi carpildiginda bize toplam urunlerin geliri gosterir.
```

```python
df['Total Revenue of Each Product Sold'] = df.apply(lambda x:x["Price Each"]*x["Total Revenue of Each Product"],axis=1)
#burda axis =1 atamamiz lazim defaultu 0 dir. eger 1 atamazsak row da price each i arayacak ve bulamayinca hata verecek
df
```

q: Which Product has the highest revenue and its number?

```python
df.groupby("Product")['Total Revenue of Each Product Sold'].mean().sort_values(ascending= False)
```

q: Find the Percent of Product Sold by Total Revenue and What is the top 5 products?

```python
df['% of Product Sold by Total Revenue']=df['Total Revenue of Each Product Sold'] / df.\
                                         groupby(['Product'])['Total Revenue of Each Product Sold'].mean().sum()*100
df.head()
```

```python
df.groupby(['Product'])['% of Product Sold by Total Revenue'].mean().sort_values(ascending=False).round(2)
```
